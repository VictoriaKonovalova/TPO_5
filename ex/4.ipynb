## Найти какое-либо изображение в сети Интернет. Создать ноутбук, в котором будет отображено выбранное изображение средствами библиотеки matplotlib по URL из сети Интернет.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from PIL import Image
import requests
```

```python
from io import BytesIO
```

```python
plt.figure(figsize=(20, 20))

url = requests.get('https://chudo-prirody.com/uploads/posts/2021-08/thumbs/1628693056_39-p-novogodnii-kotik-foto-50.jpg?size=807x807&quality=96&sign=58d30bb4711f37955435eb2d8528a11d&type=album')
img = Image.open(BytesIO(url.content))

plt.imshow(img);
```
